```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 1. DATASET: Small sample of reviews (Text into Numbers)
data = {
    'review': [
        'I love this movie, it was amazing', 'The best film I have seen',
        'Exceptional acting and great plot', 'I really enjoyed it',
        'Waste of time, terrible movie', 'I hated the acting, it was boring',
        'Worst film ever, do not watch', 'Very disappointed with the story'
    ],
    'sentiment': [1, 1, 1, 1, 0, 0, 0, 0]  # 1 = Positive, 0 = Negative
}

df = pd.DataFrame(data)

# 2. VECTORIZATION: This is the "TF-IDF" from your notes (Section 4)
# It removes "Stop Words" automatically!
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['review'])
y = df['sentiment']

# 3. THE MODEL: Logistic Regression (A simple ANN with Sigmoid)
# This uses the "Gradient Descent" you studied to find the best weights.
model = LogisticRegression()
model.fit(X, y)

# 4. TEST THE BRAIN:
def predict_sentiment(new_review):
    # Transform text using the same TF-IDF "Vocabulary"
    new_vec = vectorizer.transform([new_review])
    prediction = model.predict(new_vec)

    # Using the "Sigmoid/Softmax" logic: 1 is Pos, 0 is Neg
    return "Positive" if prediction[0] == 1 else "Negative"

# --- TRY IT OUT ---
print("--- Sentiment Bot ---")
test_text = input("Enter a movie review: ")
print(f"Result: {predict_sentiment(test_text)}")
```
